```python
import random
from datasets import Dataset

def generate_hard_negatives(example, num_negatives=15,dataset=dataset_train):
    query_index = example['index']  # Assuming this is the unique identifier in your dataset
    anchor = example['title']
    positive = example['abstract']
    
    # List to store the hard negatives
    negatives = []
    
    # Get the indexes of all other papers except the one in the query
    all_indexes = dataset_train['index']
    all_indexes.remove(query_index)  # Remove the query paper itself
    
    # Randomly sample 15 abstracts from other papers
    negative_indexes = random.sample(all_indexes, num_negatives)
    
    for idx in negative_indexes:
        negatives.append(dataset[idx]['abstract'])
    
    return {
        "query": anchor,
        "positive": positive,
        "negatives": negatives
    }
```

```python
processed_data_train = dataset_train.map(generate_hard_negatives, remove_columns=dataset_train.column_names)
```
